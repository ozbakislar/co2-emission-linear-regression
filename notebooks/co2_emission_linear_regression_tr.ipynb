# 1. Veri Seti Hakkında



Bu veri seti çeşitli araçların karbondioksit (CO₂) emisyonlarını ve yakıt tüketimini içeriyor. Makine öğrenimi bağlamında, bu veri seti karbondioksit emisyonlarını tahmin etmek veya araçların yakıt verimliliğini analiz etmek için kullanılır. Amacımız, araç özelliklerine dayanarak CO₂ emisyonlarını veya yakıt tüketimini tahmin etmektir. Toplam 7385 satır ve 12 sütun bulunmaktadır.

**Veri setindeki sütunlar ve açıklamaları:**

1. **Make**: Araç markası
2. **Model**: Araç modeli
3. **Vehicle Class**: Araç sınıfı
4. **Engine Size(L)**: Motor hacmi
5. **Cylinders**: Silindir sayısı
6. **Transmission**: Şanzıman türü
7. **Fuel Type**: Yakıt türü
8. **Fuel Consumption City (L/100 km)**: Şehir içi yakıt tüketimi
9. **Fuel Consumption Hwy (L/100 km)**: Şehir dışı (karayolu) yakıt tüketimi
10. **Fuel Consumption Comb (L/100 km)**: Kombine (şehir içi ve dışı) yakıt tüketimi
11. **Fuel Consumption Comb (mpg)**: Kombine yakıt tüketimi
12. **CO2 Emissions(g/km)**: CO₂ emisyonları

**Model**

- `4WD/4X4`: Dört tekerlekten çekiş
- `AWD`: Tüm tekerleklerden çekiş
- `FFV`: Esnek yakıtlı araç
- `SWB`: Kısa aks mesafesi
- `LWB`: Uzun aks mesafesi
- `EWB`: Genişletilmiş aks mesafesi

**Transmission**

- `A`: Otomatik
- `AM`: Otomatikleştirilmiş manuel
- `AS`: Seçmeli vitesli otomatik
- `AV`: Sürekli değişken
- `M`: Manuel
- `3 - 10`: Şanzımanın vites sayısı

**Fuel Type**

- `X`: Normal benzin
- `Z`: Premium benzin
- `D`: Dizel
- `E`: Etanol (E85)
- `N`: Doğal gaz

**Vehicle Class**

- `COMPACT`: Küçük boyutlu araçlar
- `SUV-SMALL`: Küçük boyutlu spor amaçlı araçlar
- `MID-SIZE`: Orta boyutlu araçlar
- `TWO-SEATER`: İki koltuklu araçlar
- `MINICOMPACT`: Çok küçük boyutlu araçlar
- `SUBCOMPACT`: Kompaktın altında boyutta olan küçük araçlar
- `FULL-SIZE`: Büyük boyutlu araçlar
- `STATION WAGON-SMALL`: Küçük boyutlu station wagon tipi araçlar
- `SUV-STANDARD`: Standart boyutlu spor amaçlı araçlar
- `VAN-CARGO`: Kargo için tasarlanmış vanlar
- `VAN-PASSENGER`: Yolcu taşımacılığı için tasarlanmış vanlar
- `PICKUP TRUCK-STANDARD`: Standart boyutlu pikap kamyonlar
- `MINIVAN`: Küçük boyutlu vanlar
- `SPECIAL PURPOSE VEHICLE`: Özel amaçlı araçlar
- `STATION WAGON-MID-SIZE`: Orta boyutlu station wagon tipi araçlar
- `PICKUP TRUCK-SMALL`: Küçük boyutlu pikap kamyonlar

Bu çalışmada kullanılan veri seti: https://www.kaggle.com/datasets/debajyotipodder/CO2-emission-by-vehicles/data

```python
from sklearn.linear_model import LinearRegression  
```

```python
slr.coef_
```

```python
y_pred = slr.predict(X_test)
```

```python
sns.regplot(x=y_test, y=y_pred, ci=None)

plt.xlabel('Gerçek Değerler (y Test)')
plt.ylabel('Tahmin Edilen Değerler (y Pred)')

plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def train_val(y_train, y_train_pred, y_test, y_pred, model_name):
    scores = {model_name + '_train': {'R2': r2_score(y_train, y_train_pred),
    'mae': mean_absolute_error(y_train, y_train_pred),
    'mse': mean_squared_error(y_train, y_train_pred),                          
    'rmse': np.sqrt(mean_squared_error(y_train, y_train_pred))},
    model_name + '_test': {'R2': r2_score(y_test, y_pred),
    'mae': mean_absolute_error(y_test, y_pred),
    'mse': mean_squared_error(y_test, y_pred),
    'rmse': np.sqrt(mean_squared_error(y_test, y_pred))}}
    return pd.DataFrame(scores)
```

```python
slr_score = train_val(y_train, y_train_pred, y_test, y_pred, 's_linear')

slr_score 
```

```python
avg_em = df['co2'].mean()

avg_em
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

rmse
```

```python
# Her yaptığımız tahminde yaklaşık %12 oranında hata bulunmaktadır.

rmse / avg_em
```

```python
sns.displot((y_test - y_pred), bins=50, kde=True)

plt.xlabel('Artıklar')
plt.ylabel('Frekans')

plt.show()
```
