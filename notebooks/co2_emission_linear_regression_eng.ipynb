# 1. About the Dataset



This dataset contains carbon dioxide (CO₂) emissions and fuel consumption of various vehicles. In the context of machine learning, this dataset is used to predict carbon dioxide emissions or to analyze the fuel efficiency of vehicles. Our goal is to predict CO₂ emissions or fuel consumption based on vehicle characteristics. There are a total of 7,385 rows and 12 columns.

**Columns in the dataset and their descriptions:**

1. **Make**: Vehicle brand  
2. **Model**: Vehicle model  
3. **Vehicle Class**: Vehicle class  
4. **Engine Size(L)**: Engine size  
5. **Cylinders**: Number of cylinders  
6. **Transmission**: Transmission type  
7. **Fuel Type**: Fuel type  
8. **Fuel Consumption City (L/100 km)**: City fuel consumption  
9. **Fuel Consumption Hwy (L/100 km)**: Highway fuel consumption  
10. **Fuel Consumption Comb (L/100 km)**: Combined (city and highway) fuel consumption  
11. **Fuel Consumption Comb (mpg)**: Combined fuel consumption  
12. **CO2 Emissions(g/km)**: CO₂ emissions  

**Model**

- `4WD/4X4`: Four-wheel drive  
- `AWD`: All-wheel drive  
- `FFV`: Flexible fuel vehicle  
- `SWB`: Short wheelbase  
- `LWB`: Long wheelbase  
- `EWB`: Extended wheelbase  

**Transmission**

- `A`: Automatic  
- `AM`: Automated manual  
- `AS`: Select shift automatic  
- `AV`: Continuously variable  
- `M`: Manual  
- `3 - 10`: Number of gears in the transmission  

**Fuel Type**

- `X`: Regular gasoline  
- `Z`: Premium gasoline  
- `D`: Diesel  
- `E`: Ethanol (E85)  
- `N`: Natural gas  

**Vehicle Class**

- `COMPACT`: Small-sized vehicles  
- `SUV-SMALL`: Small-sized sport utility vehicles  
- `MID-SIZE`: Mid-sized vehicles  
- `TWO-SEATER`: Two-seater vehicles  
- `MINICOMPACT`: Very small-sized vehicles  
- `SUBCOMPACT`: Small vehicles below compact size  
- `FULL-SIZE`: Large-sized vehicles  
- `STATION WAGON-SMALL`: Small-sized station wagon vehicles  
- `SUV-STANDARD`: Standard-sized sport utility vehicles  
- `VAN-CARGO`: Vans designed for cargo  
- `VAN-PASSENGER`: Vans designed for passenger transportation  
- `PICKUP TRUCK-STANDARD`: Standard-sized pickup trucks  
- `MINIVAN`: Small-sized vans  
- `SPECIAL PURPOSE VEHICLE`: Special purpose vehicles  
- `STATION WAGON-MID-SIZE`: Mid-sized station wagon vehicles  
- `PICKUP TRUCK-SMALL`: Small-sized pickup trucks  

This dataset used in this study: https://www.kaggle.com/datasets/debajyotipodder/CO2-emission-by-vehicles/data

```python
X = df[['engine_size']] 
y = df['co2']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print('Train features shape:', X_train.shape)
print('Train target shape:', y_train.shape)
print('Test features shape:', X_test.shape)
print('Test target shape:', y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression  
```

```python
slr = LinearRegression()
```

```python
slr.fit(X_train, y_train) 
```

```python
slr.coef_
```

```python
slr.intercept_
```

```python
y_train_pred = slr.predict(X_train)
```

```python
y_pred = slr.predict(X_test)
```

```python
sns.regplot(x=y_test, y=y_pred, ci=None)

plt.xlabel('Actual Values (y Test)')
plt.ylabel('Predicted Values (y Pred)')

plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def train_val(y_train, y_train_pred, y_test, y_pred, model_name):
    scores = {model_name + '_train': {'R2': r2_score(y_train, y_train_pred),
    'mae': mean_absolute_error(y_train, y_train_pred),
    'mse': mean_squared_error(y_train, y_train_pred),                          
    'rmse': np.sqrt(mean_squared_error(y_train, y_train_pred))},
    model_name + '_test': {'R2': r2_score(y_test, y_pred),
    'mae': mean_absolute_error(y_test, y_pred),
    'mse': mean_squared_error(y_test, y_pred),
    'rmse': np.sqrt(mean_squared_error(y_test, y_pred))}}
    return pd.DataFrame(scores)
```

```python
slr_score = train_val(y_train, y_train_pred, y_test, y_pred, 's_linear')

slr_score 
```

```python
avg_em = df['co2'].mean()

avg_em
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

rmse
```

```python
# There is approximately a 12% error in each prediction we make.

rmse / avg_em
```

```python
sns.displot((y_test - y_pred), bins=50, kde=True)

plt.xlabel('Residuals')
plt.ylabel('Frequency')

plt.show()
```

```python
from yellowbrick.regressor import PredictionError
```

```python
model = LinearRegression()
visualizer = PredictionError(model)

visualizer.fit(X_train, y_train)    
visualizer.score(X_test, y_test)

visualizer.show();
```
